```python
import pandas as pd
import numpy as np
import warnings, re, joblib

warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('1.scrape/cars_data.csv')
```

```python
def convert_kms(val):
    if isinstance(val, str):
        val = val.lower()
        if 'k' in val:
            return float(val.replace('k', '')) * 1000
        elif 'l' in val:
            return float(val.replace('l', '')) * 100000
    return val

def convert_price(val):
    if isinstance(val, str):
        clean_val = re.sub(r'[^\d.]', '', val)
        
        if clean_val == '':
            return 0.0
            
        if 'lakh' in val:
            return float(clean_val) * 100000
        return float(clean_val)
    return val
```

```python
from datetime import date
df['Car_Age'] = date.today().year - df['Year']
```

```python
df['Kilometers'] = df['Kilometers'].apply(convert_kms)
```

```python
df['Price'] = df['Price'].apply(convert_price)
```

```python
df.drop('City Code', axis= 1, inplace=True)
df.drop('Year', axis= 1, inplace=True)
```

```python
fuel_mapping = {'Petrol': 0, 'Diesel': 1}
df['Fuel'] = df['Fuel'].map(fuel_mapping)


# Not handle NULL values
# df['Fuel'] = df['Fuel'].replace({'Petrol': 0, 'Diesel': 1})
```

```python
trans_mapping = {'Manual': 0, 'Auto': 1}
df['Transmission'] = df['Transmission'].map(trans_mapping)
```

```python
df.dropna(subset=["Fuel"], inplace=True)
```

```python
df['Fuel'] = df['Fuel'].astype(int)
df['Transmission'] = df['Transmission'].astype(int)
df['Kilometers'] = df['Kilometers'].astype(float)
```

```python
df['Model'] = df['Model'].str.lower().str.strip()
```

```python
df.sort_values(by='Car', inplace=True)
```

```python
df = df.groupby(['Car', 'Model']).filter(lambda x: len(x) > 3)
```

```python
df.drop(df[df["Kilometers"] == 0].index, inplace=True)
df.drop(df[df["Kilometers"] == 1].index, inplace=True)
```

```python
df.to_csv("clean_cars.csv", index=False)
```
